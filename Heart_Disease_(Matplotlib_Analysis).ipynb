

#Visualisation et Analyse Exploratoire des Données


Les visualisations seront pour comprendre les variables ainsi que les
relations.


```python
# Histogramme - Variable choisie: age

import matplotlib.pyplot as plt
data['age'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title("Répartition des âges des patients")
plt.xlabel("Âge")
plt.ylabel("Nombre de patients")
plt.show()

# Type de données: Quantitative continue. L'objectif c'était de comprendre la répartition des âges des patients.
```

```python
# Diagramme à barres - Variable choisie: sex

data['sex'].value_counts().plot(kind='bar', color='orange')
plt.title("Répartition par sexe")
plt.xlabel("Sexe (0: Femme, 1: Homme)")
plt.xticks(rotation=360)
plt.ylabel("Nombre de patients")
plt.show()

# Type de données: Qualitative nominale et je voudrais visualiser la distribution des sexes par le nombre de patients
```

```python
# Diagramme circulaire (Pie Chart)- Variable choisie: target


label = ('Homme', 'Femme')
sizes = data['target'].value_counts()
colors=['lightgreen', 'salmon']
explode = (0, 0.1)
plt.pie(
    sizes,
    labels = label,
    autopct='%1.1f%%', #Affiche les pourcentages avec un chiffre après la virgule
    startangle=90,
    colors=colors,
    explode=explode,
    shadow=True
)
plt.title("Proportion de patients par sexe")
plt.ylabel("")
plt.show()

# Type de données: Qualitative nominale et l'objectif c'est de montrer la proportion de patients par sexe (proportion de patients atteints ou non de maladie cardiaque).
```

```python
# Diagramme à moustaches (Boxplot) - Variable choisie: cholesterol

data.boxplot(column='chol', by='target', grid=False)
plt.title("Cholestérol par catégorie de cible")
plt.suptitle("")  # Supprimer le titre automatique
plt.xlabel("Présence de maladie cardiaque (0: Non, 1: Oui)")
plt.ylabel("Taux de cholestérol (mg/dL)")
plt.show()

# Type de données: Quantitative continue. Je voudrais détecter les valeurs aberrantes dans le taux de cholestérol.
```

```python
# Nuage de points (Scatter Plot) - Variables choisies: age et max_heart_rate (thalach)

plt.scatter(data['age'], data['thalach'], color='purple', alpha=0.7)
plt.title("Âge vs Fréquence cardiaque maximale")
plt.xlabel("Âge")
plt.ylabel("Fréquence cardiaque maximale")
plt.show()


# Données: Quantitative continue et objectif d'identifier les corrélations entre l'âge et la fréquence cardiaque maximale.
```

```python
# Graphique en lignes - Variable choisie: resting_blood_pressure

data['trestbps'].plot(kind='line', color='red')
plt.title("Tendance de la pression artérielle au repos")
plt.xlabel("Index")
plt.ylabel("Pression artérielle au repos (mmHg)")
plt.show()

# Données: Quantitative continue, Objectif: Étudier la tendance générale de la pression artérielle au repos.
```

Afficher plusieurs graphiques en un seul coup d’œil


```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Histogramme
axes[0, 0].hist(data['age'], bins=20, color='skyblue', edgecolor='black')
axes[0, 0].set_title("Âge des patients")

# Diagramme à barres
axes[0, 1].bar(data['sex'].value_counts().index, data['sex'].value_counts().values, color='orange')
axes[0, 1].set_title("Répartition par sexe")

# Pie chart
data['target'].value_counts().plot.pie(ax=axes[1, 0], autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'salmon'])
axes[1, 0].set_title("Proportion de patients par sexe")

# Boxplot
data.boxplot(column='chol', by='target', ax=axes[1, 1], grid=False)
axes[1, 1].set_title("Cholestérol par catégorie")
axes[1, 1].set_xlabel("Target")

plt.tight_layout()
plt.show()
```

```python
# Sauvegarder les graphiques

fig = plt.figure()
data['age'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title("Répartition des âges des patients")
plt.xlabel("Âge")
plt.ylabel("Nombre de patients")
fig.savefig("distribution_age.png")
```

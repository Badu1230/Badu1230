```python
# Manipulation des données
import numpy as  np
import pandas as pd

# Visualisation
from matplotlib import cm
import matplotlib.pyplot as plt


```

```python

separateur = ","
path = "/heart.csv"

# Lire le fichier CSV en utilisant pandas
data = pd.read_csv(path, sep=separateur)

data.head(10)
```

```python
import matplotlib.pyplot as plt

# Créer un histogramme de l'âge des patients
plt.hist(data['age'], bins=range(20, 90, 10), color='skyblue', edgecolor='black')

# Ajouter des etiquettes et un titre
plt.xlabel('Âge')
plt.ylabel('Nombre de patients')
plt.title('Distribution des âges des patients')

# Afficher le graphique
plt.show()
```

```python
import matplotlib.pyplot as plt

# Créer un diagramme en barres des types de douleur thoracique en les triant par ordre
ordered_cp = data['cp'].value_counts().sort_index()
ordered_cp.plot(kind='bar', color=['skyblue', 'salmon', 'lightgreen', 'orange'])

# Ajouter des étiquettes et un titre
plt.xlabel('Type de douleur thoracique')
plt.ylabel('Nombre de patients')
plt.title('Répartition des types de douleur thoracique')
plt.xticks(rotation = 0)

# Afficher le graphique
plt.show()
```

```python
 # Créer un diagramme circulaire du genre des patients

 gender_target_yes = data[data['target'] == 'yes']

 #gender_counts = gender_target_yes.value_counts()
 #print ("gender_counts = ", gender_counts)
 gender_counts = data['sex'].value_counts()

 plt.pie(gender_counts, labels=gender_counts.index, colors=['skyblue', 'salmon'], autopct='%1.1f%%')

 # Ajouter un titre
 plt.title('Répartition du genre des patients')

 # Afficher le graphique
 plt.show()
```

```python
# Créer un diagramme en nuage de points
plt.figure(figsize=(10,6))
#data = data[data['age']<100]
plt.scatter(data['age'], data['chol'], color='blue', alpha=0.6, edgecolors='w', s=100)

# Ajouter des étiquettes et un titre
plt.xlabel('Âge')
plt.ylabel('Taux de cholestérol')
plt.title('Relation entre l\'âge et le taux de cholestérol chez les patients')
plt.grid(True)

# Afficher le graphique
plt.show()
```

```python
import matplotlib.pyplot as plt

# Extract relevant columns
age = data['Âge']
thalach = data['thalach']

# Create the line chart
plt.figure(figsize=(10, 6))
plt.plot(age, thalach, marker='o', linestyle='-', color='b', label='Max Heart Rate')

# Add labels and title
plt.xlabel("Age (years)", fontsize=12)
plt.ylabel("Maximum Heart Rate (bpm)", fontsize=12)
plt.title("Relationship Between Age and Maximum Heart Rate", fontsize=14)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
